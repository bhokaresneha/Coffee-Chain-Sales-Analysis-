## Understanding the Dataset

**Key Columns:**
- **Date**: Date of transaction
- **Market**: Market/region where sale occurred
- **Market Size**: Size category of the market (Major Market, Small Market)
- **Product**: Product category sold
- **Product Type**: Specific type of product
- **Sales**: Sales amount in dollars
- **COGS** (Cost of Goods Sold): Cost to produce/acquire the product
- **Total Expenses**: Total operational expenses
- **Marketing**: Marketing expenses
- **Profit**: Profit earned from the sale
- **Budget Profit**: Budgeted/expected profit
- **Budget COGS**: Budgeted cost of goods
- **Budget Margin**: Expected profit margin
- **Budget Sales**: Target sales amount
- **Area Code**: Geographic area code
- **State**: State where sale occurred

### Question 5: Market Analysis

### Question 7: Profit Margin Analysis

**Task:** Analyze profitability:

a) Calculate the overall profit margin percentage: (Total Profit / Total Sales) × 100

b) What is the average profit per transaction?

c) How many transactions resulted in a loss (negative profit)?

d) What percentage of transactions were profitable?

e) Which product category has the highest total profit?

f) Calculate the average profit margin for each product category

g) Create a bar chart showing total profit by product category


```python
# a
profit_margin=(coffee["Profit"].sum()/coffee["Sales"].sum())*100
print("Overall Profit Margin :",profit_margin)

# b
print("Average Profit :",coffee["Profit"].mean())

# c
loss=(coffee["Profit"]<0).sum()
print("Loss Transactions :",loss)

# d
profit_percent=((coffee["Profit"]>0).sum()/len(coffee))*100
print("Profitable Transactions :",profit_percent)

# e
product_profit=coffee.groupby("Product")["Profit"].sum().sort_values(ascending=False)
print(product_profit)

# f
coffee["Profit Margin"]=coffee["Profit"]/coffee["Sales"]*100

margin=coffee.groupby("Product")["Profit Margin"].mean()
print(margin)

# g
product_profit.plot(kind='bar')
plt.title("Profit by Product")
plt.ylabel("Profit")
plt.show()
```

### Question 9: Budget vs Actual Performance

**Task:** Compare actual performance with budgeted targets:

a) Calculate the total actual sales vs total budget sales

b) Did the company meet its overall sales budget? By how much (difference)?

c) Calculate the percentage achievement: (Actual Sales / Budget Sales) × 100

d) Calculate total actual profit vs total budget profit

e) Did the company meet its profit budget?

f) How many individual transactions exceeded their budget sales target?

g) What percentage of transactions met or exceeded their sales budget?


```python
# a
actual_sales=coffee["Sales"].sum()
budget_sales=coffee["Budget Sales"].sum()

print("\n Actual :",actual_sales)
print("\n Budget :",budget_sales)

# b
difference= actual_sales-budget_sales
print("\n Difference :",difference)

# c
achievement=(actual_sales/budget_sales)*100
print("\n Achievement % :",achievement)

# d
actual_profit=coffee["Profit"].sum()
budget_profit=coffee["Budget Profit"].sum()

print("\n Actual Profit :",actual_profit)
print("Budget Profit :",budget_profit)

# e
if actual_profit>=budget_profit:
    print("\n Profit Budget Achieved")
else:
    print("\n Profit Budget Not Achieved")

# f
exceeded=(coffee["Sales"]>=coffee["Budget Sales"]).sum()
print("\n Exceeded Budget :",exceeded)

# g
percentage=(exceeded/len(coffee))*100
print("\n Percentage :",percentage)
```

### Question 10: Product Type Deep Dive

**Task:** Analyze specific product types in detail:

a) How many unique product types are available?

b) List all product types

c) For each product type, calculate:
   - Total sales
   - Total profit
   - Number of transactions
   - Average sales per transaction

d) Which product type generates the most revenue?

e) Which product type is the most profitable?

f) Create a summary table showing the top 10 product types by total sales

g) Visualize the top 10 product types using a horizontal bar chart


```python
#a
print("\n",coffee["Product Type"].nunique()," unique product types ")

#b
product_type=coffee["Product Type"].unique()
print("\n Product Types:",product_type)

#c
summary=coffee.groupby("Product Type").agg(
    Total_Sales=("Sales","sum"),
    Total_Profit=("Profit","sum"),
    Transactions=("Product Type","count"),
    Average_Sales=("Sales","mean")
).sort_values(by="Total_Sales",ascending=False)
print("\n",summary)

# d
print("\n Highest Revenue Product Type")
print(summary["Total_Sales"].idxmax())

# e
print("\n Most Profitable Product Type")
print(summary["Total_Profit"].idxmax())

#f
#print(coffee.groupby("Product Type")["Sales"].sum().head(10))
print(summary["Total_Sales"].head(10))

#g
summary["Total_Sales"].head(10).plot(
    kind="barh",
    y="Total_Sales",
    legend=False
)
plt.title("Top 10 Product Types by Sales")
plt.xlabel("Sales")
plt.show()
```
